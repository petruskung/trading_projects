```python
# Group Anagrams
# 99th percentile for Time; 80th percentile for Memory
class Solution(object):
    def groupAnagrams(self, strs):
        grouped_anagram = {}
        for string in strs:
            sorted_string = "".join(sorted(string))
            if sorted_string in grouped_anagram:
                grouped_anagram[sorted_string].append(string)
            else:
                grouped_anagram[sorted_string] = [string]
        return list(grouped_anagram.values())
```

```python
# Top K Frequent Elements
# 99th percentile for Time; 88th percentile for Memory
class Solution(object):
    def topKFrequent(self, nums, k):
        number_with_count = {}
        for i in nums:
            if i in number_with_count:
                number_with_count[i] += 1
            else:
                number_with_count[i] = 1
        top_k_keys = list(sorted(number_with_count, key=number_with_count.get, reverse=True)[:k])
        return top_k_keys
```
